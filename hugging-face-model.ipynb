```python
import pandas as pd
from simpletransformers.ner import NERModel, NERArgs
from sklearn.model_selection import train_test_split
```

```python
train_path = '../input/scl-2021-ds/train.csv'
df = pd.read_csv(train_path)
df
```

```python
df['POI'] = df['POI/street'].apply(lambda x: x.split('/')[0])
df['street'] = df['POI/street'].apply(lambda x: x.split('/')[1])
```

```python
df
```

```python
no_nan_train_df = df.replace({'': pd.NA}).dropna()
no_nan_train_df
```

```python
no_nan_train_df['address_no_punct'] = no_nan_train_df['raw_address'].str.replace('[^\w\s]','')
no_nan_train_df['address_no_punct']
```

```python
train_df, test_df = train_test_split(no_nan_train_df, test_size = 0.1)
```

```python
train_df
```

```python
test_df
```

```python
list_id = []
list_word = []
list_tag = []
```

```python
def insert_into_list(address_id, address, poi, street):
    poi_list = poi.split()
    street_list = street.split()
    for i in address.split():
        list_word.append(i)
        list_id.append(address_id)
        if i in poi_list:
            list_tag.append('POI')
        elif i in street_list:
            list_tag.append('street')
        else:
            list_tag.append('O')
```

```python
def generate_list(df_to_list):
    list_id = []
    list_word = []
    list_tag = []
    for idx, row in df_to_list.iterrows():
        insert_into_list(row['id'], row['address_no_punct'], row['POI'], row['street'])
```

```python
generate_list(train_df)
```

```python
list_id
```

```python
list_word
```

```python
list_tag
```

```python
ner_train_input_df = pd.DataFrame({
    'sentence_id': list_id,
    'words': list_word,
    'labels': list_tag
})
ner_train_input_df
```

```python
generate_list(test_df)
ner_test_input_df = pd.DataFrame({
    'sentence_id': list_id,
    'words': list_word,
    'labels': list_tag
})
ner_test_input_df
```

```python
model_args = NERArgs()
model_args.train_batch_size = 16
model_args.evaluate_during_training = True
tag_list = ['POI', 'street', 'O']
```

```python
model = NERModel('roberta', 'roberta-base', labels = tag_list, 
                 args = model_args, use_cuda = True)
```

```python
model.train_model(ner_train_input_df, eval_data = ner_test_input_df, output_dir = './outputs/model_output')
```

```python
validation_path = '../input/scl-2021-ds/test.csv'
validation_df = pd.read_csv(validation_path)
```

```python
validation_df
```

```python
validation_df['address_no_punct'] = validation_df['raw_address'].str.replace('[^\w\s]','')
validation_df
```

```python
pred, raw = model.predict(validation_df['address_no_punct'])
pred
```

```python
for i in pred:
    print(i)
    print()
```

```python
def process_pred(pred_row):
    POI = ''
    street = ''
    for i in pred_row:
        for j in i:
            if i[j] == 'POI':
                POI += ' ' + j
            elif i[j] == 'street':
                street += ' ' + j
    return POI[1:] + '/' + street[1:]
```

```python
res = []
for i in pred:
    res.append(process_pred(i))
```

```python
print(res)
```

```python
res_df = pd.DataFrame({
    'id': validation_df['id'],
    'POI/street': res
})
res_df
```
