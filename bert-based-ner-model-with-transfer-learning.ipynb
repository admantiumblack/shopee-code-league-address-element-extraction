```python
import pandas as pd
from simpletransformers.ner import NERModel, NERArgs
from sklearn.model_selection import train_test_split
```

```python
train_path = '../input/scl-2021-ds/train.csv'
df = pd.read_csv(train_path)
df
```

```python
df['POI'] = df['POI/street'].apply(lambda x: x.split('/')[0])
df['street'] = df['POI/street'].apply(lambda x: x.split('/')[1])
```

```python
train_df, test_df = train_test_split(df, test_size = 0.1)
```

```python
train_df
```

```python
test_df
```

```python
list_id = []
list_word = []
list_tag = []
```

```python
def insert_into_list(address_id, address, poi, street):
    poi_list = poi.split()
    street_list = street.split()
    for i in address.split():
        list_word.append(i)
        list_id.append(address_id)
        tag = ''
        if i in poi_list:
            if i == poi_list[0]:
                tag = 'B-ORG'
            else:
                tag = 'I-ORG'
        elif i in street_list:
            if i == street_list[0]:
                tag = 'B-LOC'
            else:
                tag = 'I-LOC'
        else:
            tag = 'O'
        list_tag.append(tag)
```

```python
def generate_list(df_to_list):
    list_id = []
    list_word = []
    list_tag = []
    for idx, row in df_to_list.iterrows():
        insert_into_list(row['id'], row['address_no_punct'], row['POI'], row['street'])
```

```python
generate_list(train_df)
```

```python
list_id
```

```python
list_word
```

```python
list_tag
```

```python
ner_train_input_df = pd.DataFrame({
    'sentence_id': list_id,
    'words': list_word,
    'labels': list_tag
})
ner_train_input_df
```

```python
generate_list(test_df)
ner_test_input_df = pd.DataFrame({
    'sentence_id': list_id,
    'words': list_word,
    'labels': list_tag
})
ner_test_input_df
```

```python
model_args = NERArgs()
model_args.train_batch_size = 8
model_args.evaluate_during_training = False
model_args.num_train_epochs = 1

model_args.save_steps = -1
model_args.save_optimizer_and_scheduler = False
model_args.save_model_every_epoch = False
model_args.overwrite_output_dir = True

tag_list = ["B-CRD",
            "B-DAT",
            "B-EVT",
            "B-FAC",
            "B-GPE",
            "B-LAN",
            "B-LAW",
            "B-LOC",
            "B-MON",
            "B-NOR",
            "B-ORD",
            "B-ORG",
            "B-PER",
            "B-PRC",
            "B-PRD",
            "B-QTY",
            "B-REG",
            "B-TIM",
            "B-WOA",
            "I-CRD",
            "I-DAT",
            "I-EVT",
            "I-FAC",
            "I-GPE",
            "I-LAN",
            "I-LAW",
            "I-LOC",
            "I-MON",
            "I-NOR",
            "I-ORD",
            "I-ORG",
            "I-PER",
            "I-PRC",
            "I-PRD",
            "I-QTY",
            "I-REG",
            "I-TIM",
            "I-WOA",
            "O"]
# model_args.labels_list = tag_list
```

```python
model = NERModel('xlmroberta', 'cahya/xlm-roberta-base-indonesian-NER', labels = tag_list,
                 args = model_args, use_cuda = True)
```

```python
model.train_model(ner_train_input_df, eval_data = ner_test_input_df, output_dir = './outputs/model_output')
```

```python
validation_path = '../input/scl-2021-ds/test.csv'
validation_df = pd.read_csv(validation_path)
```

```python
validation_df
```

```python
validation_df['address_no_punct'] = validation_df['raw_address'].str.replace('[^\w\s]','')
validation_df
```

```python
pred, raw = model.predict(validation_df['address_no_punct'])
pred
```

```python
# for i in pred:
#     print(i)
#     print()
```

```python
def process_pred(pred_row):
    POI = ''
    street = ''
    for i in pred_row:
        for j in i:
            if i[j] in ['B-ORG', 'I-ORG']:
                POI += ' ' + j
            elif i[j] in ['B-LOC', 'I-LOC']:
                street += ' ' + j
    return POI[1:] + '/' + street[1:]
```

```python
res = []
for i in pred:
    res.append(process_pred(i))
```
